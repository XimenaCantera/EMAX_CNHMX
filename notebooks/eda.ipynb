```python
df_mantenimiento['delay_vs_service_interval'] = df_mantenimiento['ACTUAL'] - df_mantenimiento['SERVICIO']

status_risk_mapping = {'Cerrada': 0,'PorVencer': 1,'EnProceso': 1,'Abierta': 1,'Pendiente': 2,'CerradaFuera': 2 }


df_mantenimiento['overdue_risk'] = df_mantenimiento['ESTATUS'].map(status_risk_mapping)


rit_agg_df = df_mantenimiento.groupby('ALIAS').agg( avg_overdue_risk=('overdue_risk', 'mean'),
                                      avg_delay_vs_service_interval=('delay_vs_service_interval', 'mean')).reset_index()
```

```python
print("Tabla cruzada: Riesgo vs Estatus")
estatus_Unidades = df_mantenimiento['ESTATUS'].value_counts()
riesgo_unidades = df_mantenimiento['overdue_risk'].value_counts()

pd.crosstab(df_mantenimiento['ESTATUS'], df_mantenimiento['overdue_risk'])
```

La tabla confirma la correcta asignación de niveles de riesgo a cada estado de servicio.
El patrón es que los servicios Pendientes y Cerrados Fuera tienen el mayor riesgo (2.0). Hacer esta categorización nos permite fortalecer la priorización de las unidades.
